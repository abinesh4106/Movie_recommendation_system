# **Content and Popularity based movie Recommendation System**

```python
#import dependencies

import numpy as np
import pandas as pd

# To find the close match of the movie given by the user.
import difflib

# To convert Textual data into numerical value
from sklearn.feature_extraction.text import TfidfVectorizer

# For similarity score
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('https://drive.google.com/file/d/1b7m1MMBbf9J7gV5_P4M2mDSFQIPh3AAy/view?usp=sharing')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# Feature selection for Recommendations (recommendations need some of the features only)

# For content and popularity based recommendations system

selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# Cleaning the datas in the Selected Features

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('') # replacing the Empty values in the column with null string
```

```python
# Combining the selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
# Converting the text data into feature vectors

vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# Cosine similarity Score

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
similarity.shape
```

```python
# Getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

```python
# Creating a list with all the movies in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# Find the close match of the given input

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# Find the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# Getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
# Recommend the higher similarities based on the input

#sorting based on similarities

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print("Suggested movies are :")

i = 0
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]

  if i<30:
    print(i, ',', title_from_index)
    i+=1
```
